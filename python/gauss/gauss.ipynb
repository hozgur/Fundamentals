```python
def count_ones_distribution(n):
    ones_count_dict = {}
    for i in range(1, n + 1):
        # İkilik sistemdeki temsili al ve '1' karakterlerini say
        count = bin(i).count('1')
        if count in ones_count_dict:
            ones_count_dict[count] += 1
        else:
            ones_count_dict[count] = 1
    return ones_count_dict

# Sonuçları yazdırmak için
def print_results(n):
    distribution = count_ones_distribution(n)
    for ones_count, frequency in sorted(distribution.items()):
        print(f"Birlerin Sayısı: {ones_count}, Frekans: {frequency}")

# Örnek kullanım:
print_results(16)
```

```python
import matplotlib.pyplot as plt

def count_ones_distribution(n):
    ones_count_dict = {}
    for i in range(1, n + 1):
        count = bin(i).count('1')
        if count in ones_count_dict:
            ones_count_dict[count] += 1
        else:
            ones_count_dict[count] = 1
    return ones_count_dict

def plot_distribution(n):
    distribution = count_ones_distribution(n)
    # '1' sayılarını ve frekanslarını listeler haline getir
    ones_counts = list(distribution.keys())
    frequencies = list(distribution.values())
    
    plt.figure(figsize=(10, 5))  # Grafik boyutunu ayarla
    plt.bar(ones_counts, frequencies, color='blue')  # Bir histogram çiz
    plt.xlabel('Birlerin Sayısı')  # X ekseni başlığı
    plt.ylabel('Frekans')  # Y ekseni başlığı
    plt.title(f'1\'den {n}\'e Kadar Olan Sayıların İkilik Sistemde Birlerin Frekans Dağılımı')  # Grafik başlığı
    plt.xticks(ones_counts)  # X ekseni işaretleri, her birlerin sayısı için bir işaret
    plt.show()

# Örnek kullanım:
plot_distribution(1023)
```
